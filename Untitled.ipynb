DNA sequencing is the process of determining the sequence of nucleotides within a DNA molecule. Every organism's DNA consists of a unique sequence of nucleotides. Determining the sequence can help scientists compare DNA between organisms, which can help show how the organisms are related.

```python
def getKmers(sequence, size=6):
    return [sequence[x:x+size].lower() for x in range(len(sequence) - size + 1)]
```

```python
human_data['words'] = human_data.apply(lambda x: getKmers(x['sequence']), axis=1)
human_data = human_data.drop('sequence', axis=1)
```

```python
#human_data['words'][1]
human_data.head(2)
```

#### because we want to perform bagging or tf-idf we need to change these values in sentence or paragraph

```python
human_texts = list(human_data['words'])
for item in range(len(human_texts)):
    human_texts[item] = ' '.join(human_texts[item])
y_data = human_data.iloc[:, 0].values
```

```python
y_data
```

```python
print(human_texts[0])
```

### let's perform tf_idf 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(human_texts)
```

```python
print(X.shape)
```

##### we should also try bag of words just 2 see which give better accuracy

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(2,2))
y = cv.fit_transform(human_texts)
```

```python
print(y.shape)
```

###### lets split the data 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y_data, 
                                                    test_size = 0.20, 
                                                    random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB(alpha=0.1)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
print("Confusion matrix\n")
print(pd.crosstab(pd.Series(y_test, name='Actual'), pd.Series(y_pred, name='Predicted')))
def get_metrics(y_test, y_predicted):
    accuracy = accuracy_score(y_test, y_predicted)
    precision = precision_score(y_test, y_predicted, average='weighted')
    recall = recall_score(y_test, y_predicted, average='weighted')
    f1 = f1_score(y_test, y_predicted, average='weighted')
    return accuracy, precision, recall, f1
accuracy, precision, recall, f1 = get_metrics(y_test, y_pred)
print("accuracy = %.3f \nprecision = %.3f \nrecall = %.3f \nf1 = %.3f" % (accuracy, precision, recall, f1))
```
